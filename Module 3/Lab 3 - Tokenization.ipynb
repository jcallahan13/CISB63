### complete the block of code below - (one line of code)

```python
text = "Who would have thought that computer programs would be analyzing human sentiments"

### tokenize the text and assign the result to tokens. hint: use word_tokenize

tokens = word_tokenize(text)

# end of your code

lemmatizer = WordNetLemmatizer()
tokens=[lemmatizer.lemmatize(word) for word in tokens]
```

#### import RegexpTokenizer from nltk.tokenize

```python
from nltk.tokenize import RegexpTokenizer
```

#### tokenize the followign text using RegexpTokenizer <br> include every word, number, symbols, and white spaces

```python
text = "Who would have thought that computer programs would be analyzing human sentiments? Isn't is amazin?"

tokenizer = RegexpTokenizer(r'\w+|\$[d\.]|\W+|\s+')

## end of your code
tokenizer.tokenize(text)
```

#### import BlanklineTokenizer from nltk.tokenize

```python
from nltk.tokenize import BlanklineTokenizer
```

#### tokenize the followign text using BlanklineTokenizer

```python
text = "Who would have thought that computer programs would be analyzing human sentiments?\n\n Isn't is amazin?\n\n I hope you are enjoying this class.\n\n The price of our book is $42"

tokenizer=BlanklineTokenizer()

## end of your code

tokenizer.tokenize(text)
```
